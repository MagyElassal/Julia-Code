```julia
using PyPlot
```

```julia
function LighthouseProb(N) #input number of flashes "N"                  
    #----------- Data generation------------------------
    α = 1                       # position of the lighthouse
    β= 1       
    θ = rand(-π/2:π/2,1,N) 
    
    X= β .* tan(θ) .+ α           #the vector of flashes positions
    #----------------Bayesian Calculations-----------------------------------
    α_range = [-5:0.1:5...]
    L= -sum(log(β^2 .+ (X .- α_range).^2),2)
    getMaxL= maximum(L);
    modifiedL= L .- getMaxL;
    posterior = exp(modifiedL)
    μ = mean(posterior)
    
    #--------Ploting Postrior---------------------------- 
    plot(α_range, posterior)
    xlabel("Lighthouse Position α (km)")
    ylabel("prob(α|{X},β,I)")
    
    println("Mean Value: ",μ)
    
end
```
