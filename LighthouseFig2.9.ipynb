```julia
using PyPlot
```

```julia
leighthouseProb()=(minR = -pi/2;
maxR= pi/2;
r= minR + (rand())/ (maxR- minR))
```

```julia
N=3;
t= zeros(1,N);
for k=1:N  
    t[k]=leighthouseProb();
end
beta= 1;
alfa= [-5:0.1:5];
t
```

```julia
x= beta .* tan(t) .+ alfa
```

```julia
x .- alfa
```

```julia
tan(t)
```

```julia
L= -sum(log(beta^2 .+ (x .- alfa).^2),2)
```

```julia
plot(alfa,L)
```

```julia
getMaxL= maximum(L);
modifiedL= L .- getMaxL;
plot(alfa,modifiedL)
```

```julia
posterior = exp(modifiedL)
plot(alfa, posterior)
```

```julia
testPost= exp(L)
plot(alfa,testPost)
```
