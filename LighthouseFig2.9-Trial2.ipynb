```julia
using PyPlot
```

```julia
function lighthouseProb()
    minR = -pi/2;
    maxR= pi/2;
    r= minR + rand() * (maxR - minR)
end
```

```julia
function lighthouseExamp(N)
    beta= 1;

    alfa= [-5 : 0.01 : 5];
    R= length(alfa);

   t= zeros(R,N);

    for k=1:N, i=1:R  
       t[i,k]=lighthouseProb();
    end
    #t= zeros(1,N)
    #for k=1:N  
    #    t[k]=lighthouseProb();
    #end

    x = beta .* tan(t) .+ alfa
    L= -sum(log(beta^2 .+ (x .- alfa).^2),2)
    getMaxL= maximum(L);
    modifiedL= L .- getMaxL;
    posterior = exp(modifiedL)
    println(N)
    plot(alfa, posterior)
    
    xlabel("Lighthouse Position alfa (km)")
    ylabel("prob(alfa|{x_k},beta,I)")
    
end
```
